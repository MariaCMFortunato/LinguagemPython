A classe Pessoa terá um atributo chamado nrp_total.<br>
Este atributo não precisa do self, porque o self refere-se ao objeto e não à classe<br>
O atributo da classe será definido como protegido, portanto pode ser acedido pelos objetos do tipo Pessoa e Empregado<br>
A seguir à definição dos métodos para objetos do tipo Pessoas, criamos o método da classe get_pessoas_total()

```python
class Pessoa:
#Define pessoas com nome, idade e nif
#Atributo da classe que contém o número total de pessoa - é protegido
    _nrpTotal = 0
    def __init__(self, nome, idade, nif):
        self.__nome = nome
        self.idade = idade
        self._nif = nif
        #quando se cria um objeto, aumenta o contador
        Pessoa._nrpTotal += 1
#Métodos da classe
    def getNome(self):
        return self.__nome
    #Nome não pode ser alterado porque é um atributo privado
    #def setNome(self, nome):
    #    self.nome = nome
    def getIdade(self):
        return self.idade
    def setIdade(self, idade):
        self.idade = idade
    def _getNif(self):
        return self._nif
    def _setnif(self, nif):
        self._nif = nif
    
    #método estático na classe Pessoa
    def getPessoasTotal():
        return Pessoa._nrpTotal


class Empregado(Pessoa):
#Pessoa = superclasse; Empregado = subclasse
    def __init__(self, nome, idade, nif, empresa, salario):
        super().__init__(nome, idade, nif)
        self.empresa = empresa
        self.salario = salario
    def getEmpresa(self):
        return self.empresa
    def setEmpresa(self, empresa):
        self.empresa = empresa
    def getSalario(self, salario):
        return self.salario
    def setSalario(self, salario):
        self.salario = salario
```

Os erros anteriores são devidos ao facto de que ao acedermos a uma método através de um objeto é sempre passado o argumento SELF.<br>
Para que o métodos da classe possam ser acedidos pelos objetos, é necessário tornar o método estático através do decorardor @staticmethod

```python
class Pessoa:
#Define pessoas com nome, idade e nif
#Atributo da classe que contém o número total de pessoa - é protegido
    _nrpTotal = 0
    def __init__(self, nome, idade, nif):
        self.__nome = nome
        self.idade = idade
        self._nif = nif
        #quando se cria um objeto, aumenta o contador
        Pessoa._nrpTotal += 1
#Métodos da classe
    def getNome(self):
        return self.__nome
    #Nome não pode ser alterado porque é um atributo privado
    #def setNome(self, nome):
    #    self.nome = nome
    def getIdade(self):
        return self.idade
    def setIdade(self, idade):
        self.idade = idade
    def _getNif(self):
        return self._nif
    def _setnif(self, nif):
        self._nif = nif
    
    #método estático na classe Pessoa
    @staticmethod
    def getPessoasTotal():
        return Pessoa._nrpTotal


class Empregado(Pessoa):
#Pessoa = superclasse; Empregado = subclasse
    def __init__(self, nome, idade, nif, empresa, salario):
        super().__init__(nome, idade, nif)
        self.empresa = empresa
        self.salario = salario
    def getEmpresa(self):
        return self.empresa
    def setEmpresa(self, empresa):
        self.empresa = empresa
    def getSalario(self, salario):
        return self.salario
    def setSalario(self, salario):
        self.salario = salario
```

Variáveis globais / variáveis Locais <br>
As variáveislocais são definidas e conhecidas apenas nas funções onde são utilizadas.<br> .
As variáveis globais podem ser globais por atribuição ou por definição:

```python
#Uso de variáveis globais por atribuição
def soma():
    print("adição:", a+b)
def main():
    print("multiplicação:", a*b)
    soma()
a = 2
b = 3
if __name__ == "__main__":
    main()
```

```python
#Uso de variáveis globais por definição
def soma():
    print("adição:", a + b)
def main():
    global a, b
    a = 2
    b = 3
    print("multiplicação:", a * b)
    soma()

if __name__ == "__main__":
    main()
```

```python
#Uso de variáveis locais
def soma(a, b):
    print("adição:", a + b)
def main():
    a = 2
    b = 3
    print("multiplicação:", a * b)
    soma(a, b)

if __name__ == "__main__":
    main()
```
